```python
import json

RAT = 115
TOTAL_RUN = 10

with open('data/Runs/{!s}_{!s}_Runs/result.json'.format(RAT,TOTAL_RUN), "r", encoding='utf-8') as jsonFile:
    data = json.load(jsonFile)
```

```python
from matplotlib import pyplot as plt
from pylab import rcParams

rcParams['figure.figsize'] = 20, 5
#plot
fig, ax=plt.subplots(1,2)

model = {
    'IF':{'name':'IF', 'color':'orange', 'lstyle':'solid'},
    'AE':{'name':'AE', 'color':'purple', 'lstyle':'dashed'},
    'AAE':{'name':'AAE', 'color':'green', 'lstyle':'dashdot'},
    'OCSVM':{'name':'SVDD', 'color':'blue', 'lstyle':'dotted'},
    'GAN-AE':{'name':'GAN-AE', 'color':'pink', 'lstyle':'dashdot'},
    'Proposed Method':{'name':'FinGANomaly(FE)_0.1', 'color':'lime', 'lstyle':'solid'},
    'Proposed Method':{'name':'FinGANomaly_0.1', 'color':'black', 'lstyle':'solid'}
}
#colors = ['orange','purple','green','lime','black','yellow','brown','pink','blue','lightblue']

#ROC curve
ax[0].set_title('Receiver Operating Characteristic')
for m in model:
    lbl, c, ls = model[m]['name'], model[m]['color'], model[m]['lstyle']
    try:
        fpr, tpr = data[lbl]['AVG']['FPR'], data[lbl]['AVG']['TPR']
        auc = sum(data[lbl]['AUC']) / TOTAL_RUN
        ax[0].plot(fpr, tpr, color = c, linestyle=ls, label = '{!s} = {:.4f}'.format(m, auc))
        print(lbl)
    except:
        pass
ax[0].plot([0, 1], [0, 1],'r--', label = 'baseline = {:.4f}'.format(0.5))
ax[0].legend(loc = 'lower right')
ax[0].set_xlim([0, 1])
ax[0].set_ylim([0, 1])
ax[0].set_ylabel('True Positive Rate')
ax[0].set_xlabel('False Positive Rate')

#PR curve
ax[1].set_title('Precision-Recall Curve')
for m in model:
    lbl, c, ls = model[m]['name'], model[m]['color'], model[m]['lstyle']
    try:
        r, p = data[lbl]['AVG']['R'], data[lbl]['AVG']['P']
        auprc = sum(data[lbl]['AUPRC']) / TOTAL_RUN
        ax[1].plot(r, p, color = c, linestyle=ls, label='{!s} = {:.4f}'.format(m, auprc))
        print(lbl)
    except:
        pass
proportion_of_abn =  1 / (RAT+1)
ax[1].plot([0, 1], [proportion_of_abn, proportion_of_abn],'r--', label = 'baseline = {:.4f}'.format(proportion_of_abn))
ax[1].legend(loc = 'lower left')
ax[1].set_xlim([0, 1])
ax[1].set_ylim([0, 1])
ax[1].set_ylabel('Precision')
ax[1].set_xlabel('Recall')

plt.show()
```
